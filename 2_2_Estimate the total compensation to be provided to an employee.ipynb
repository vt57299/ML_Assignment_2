```python
import pandas as pd
```

```python
df = pd.read_csv('train_set.csv')
df
```

```python
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True,drop = True)
df
```

```python
df.describe(include = 'all')
```

```python
df.isna().sum()
```

```python
df = df[df['Union'].notna()]
df = df[df['JF'].notna()]
df.reset_index(inplace = True, drop = True)
```

```python
for i in ['OG','DC','Dept','Union','JF','Job','YT']:
    print(df[i].nunique())
```

```python
# converting text data to numeric
from sklearn.preprocessing import LabelEncoder
cols = ['OG','DC','Dept','Union','JF','Job','YT']
le = LabelEncoder()
for columns in cols:
    df[columns] = le.fit_transform(df[[columns]])
```

```python
df.head()
```

```python
# checking for outliers
import seaborn as sns
sns.boxplot(df.Total_Compensation,orient = 'h')
```

```python
# dealing with outliers
df = df[df.Salaries < 380000]
df = df[df.Overtime < 175000]
df = df[df.Total_Compensation < 600000]
```

```python
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True, drop = True)
df
```

```python
x = df.drop(['Total_Compensation'], axis = 1)
y = df['Total_Compensation']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn = KNeighborsRegressor()
kn.fit(x_train,y_train)
```

```python
y_pred1 = kn.predict(x_test)
y_pred1
```

```python
print(r2_score(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
```
